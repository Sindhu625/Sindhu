```python
#Create a 1D array
import numpy as np
arr = ([1,2,3,4,5])
print(arr)
```

```python
#Create a 2D array
import numpy as np
arr = ([1,2,3],
       [4,5,6])
print(arr)
```

```python
#Create an array of zeros
zeros = np.zeros((3, 3))
print(zeros)
```

```python
#Create an array of ones
ones = np.ones((2,3))
print(ones)
```

```python
#Create a range array
r=np.arange(0,10,2)
print(r)
```

```python
#Random array
rand=np.random.rand(1,3)
print(rand)
```

```python
#Arithmetic operations
a=np.array([1,3,2])
b=np.array([4,6,5])
print(a+b)
print(a-b)
print(a*b)
print(a/b)
```

```python
#Max,Min,Sum
arr=np.array([10,20,5,15])
print(arr.max())
print(arr.min())
print(arr.sum())
```

```python
#Slicing
arr=np.array([10,20,30,40,50])
print(arr[1:4])
```
